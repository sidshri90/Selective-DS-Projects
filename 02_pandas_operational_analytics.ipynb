```python
# Cell 2: Raw Operational Data Simulation
# Creating a dictionary with intentional formatting discrepancies and missing values

raw_log_data = {
    'Transaction_Timestamp': ['2026-07-09 08:30:00', '2026-07-09 09:15:00', None, '2026-07-09 11:45:00', '2026-07-09 14:20:00'],
    'Territory_ID': ['PAT-01', 'PAT-02', 'PAT-01', 'PAT-03', 'PAT-02'],
    'Operational_Value': [12500.0, None, 14200.0, 9800.0, 15500.0],
    'Log_Status': [' SUCCESS ', ' PENDING ', ' SUCCESS', 'FAILED', ' SUCCESS ']
}

# Convert the dictionary into an enterprise Pandas DataFrame
df_log = pd.DataFrame(raw_log_data)

print("== STEP 3: PIPELINE INGESTION COMPLETE ==")
df_log
```

```python
# Cell 3: Column Vector String Cleansing
# Stripping whitespace from text boundaries to ensure exact string matching

print("== BEFORE SANITIZATION (Raw String Lengths) ==")
print(df_log['Log_Status'].tolist())

# Apply the string strip transformation directly to the vector
df_log['Log_Status'] = df_log['Log_Status'].str.strip()

print("\n== AFTER SANITIZATION (Sanitized String Lengths) ==")
print(df_log['Log_Status'].tolist())
```

```python
# Cell 3: Column Vector String Cleansing
# Stripping whitespace from text boundaries to ensure exact string matching

print("== BEFORE SANITIZATION (Raw String Lengths) ==")
print(df_log['Log_Status'].tolist())

# Apply the string strip transformation directly to the vector
df_log['Log_Status'] = df_log['Log_Status'].str.strip()

print("\n== AFTER SANITIZATION (Sanitized String Lengths) ==")
print(df_log['Log_Status'].tolist())
```

```python
# Cell 4: Statistical Imputation of Missing Numeric Variables
# Calculating the median of the column and using it to fill missing data points

# Calculate the median value (ignoring NaN elements)
median_value = df_log['Operational_Value'].median()
print(f"Calculated Operational Column Median: {median_value}")

# Fill the missing values using .fillna()
df_log['Operational_Value'] = df_log['Operational_Value'].fillna(median_value)

print("\n== STEP 5: IMPUTATION COMPLETE ==")
df_log
```

```python
# Cell 5: Operational Group Aggregations
# Grouping by final log states to extract business metrics and cumulative sums

print("== STEP 6: EXECUTING METRIC EXTRACTION ==")

# Calculate sum and mean simultaneously using descriptive aggregations
performance_metrics = df_log.groupby('Log_Status')['Operational_Value'].agg(['count', 'sum', 'mean']).round(2)

print("\n== FINAL OPERATIONAL METRICS FOR EXECUTIVE REPORTING ==")
print(performance_metrics)
```

```python
# Cell 10: Logistics High-Volume DateTime Ingestion Pipeline
import pandas as pd
import io

# Simulating a raw flight-delays logistics data stream with missing intervals and string variations
logistics_raw_data = """Carrier,Flight_Num,Scheduled_Departure,Actual_Departure,Delay_Reason
Indigo,IND-402,2026-07-10 13:00:00,2026-07-10 13:15:00,Weather
AirIndia,AI-182,2026-07-10 14:30:00,None,Gate Delay
SpiceJet,SG-901,2026-07-10 15:00:00,2026-07-10 15:00:00,None
Indigo,IND-211,2026-07-10 18:45:00,2026-07-10 19:22:00,Carrier Delay
AirIndia,AI-884,2026-07-10 21:00:00,2026-07-10 21:05:00,None
"""

# Load dataset and explicitly force Pandas to recognize structural components
df_logistics = pd.read_csv(io.StringIO(logistics_raw_data))

print("== LOGISTICS FLIGHT DATASTREAM INGESTED ==")
df_logistics
```

```python
# Cell 11: Vectorized Datetime Transformations and Metric Derivation
# Converting text series to datetime objects and extracting exact operational delay differentials

print("== STEP 11: RUNNING OPERATIONAL DATETIME MATH ==")

# 1. Convert text strings into formal Pandas Datetime objects
df_logistics['Scheduled_Departure'] = pd.to_datetime(df_logistics['Scheduled_Departure'])
df_logistics['Actual_Departure'] = pd.to_datetime(df_logistics['Actual_Departure'])

# 2. Calculate the exact delta in minutes using vectorized subtraction
# .dt.total_seconds() / 60 isolates the exact floating-point minute variance
df_logistics['Delay_Minutes'] = (df_logistics['Actual_Departure'] - df_logistics['Scheduled_Departure']).dt.total_seconds() / 60

# 3. For flights that never departed (NaN), handle the missing delta cleanly by defaulting to 0.0 for calculations
df_logistics['Delay_Minutes'] = df_logistics['Delay_Minutes'].fillna(0.0)

print("\n== LOGISTICS TIMELINE TRANSFORMATIONS COMPLETE ==")
df_logistics[['Carrier', 'Flight_Num', 'Scheduled_Departure', 'Actual_Departure', 'Delay_Minutes', 'Delay_Reason']]
```

```python
# Cell 12: Multi-Variable Carrier Performance Metrics
# Aggregating logistics volume and calculating metrics for operational reporting

print("== STEP 12: COMPILING CARRIER SCORECARD ==")

# Group by Carrier to isolate total flight counts, maximum delays, and sum of total delay minutes
carrier_scorecard = df_logistics.groupby('Carrier')['Delay_Minutes'].agg(
    Total_Flights='count',
    Max_Delay_Minutes='max',
    Total_Delay_Minutes='sum'
).reset_index()

print("\n== EXECUTIVE LOGISTICS REPORT ==")
carrier_scorecard
```

```python
# Cell 13: Staging Core Aggregations to Disk Storage
# Exporting the clean carrier analytics matrix for downstream database ingestion

print("== STEP 13: EXPORTING LOGISTICS METRICS TO DISK ==")

# Define our target destination file name
logistics_export_file = "carrier_performance_scorecard.csv"

# Write out the aggregated dataframe to a physical CSV file
carrier_scorecard.to_csv(logistics_export_file, index=False)

print(f"\nSUCCESS: Operational report successfully compiled and saved as: '{logistics_export_file}'")
print("Your local file repository is now updated with this live report.")
```
